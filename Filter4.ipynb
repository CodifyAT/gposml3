```python
import pandas as pd
df=pd.read_csv('gsol4.csv')
df.head()
```

```python
del df["Unnamed: 0"]
```

```python
from statistics import mean,median,mode
plhh=[]
sbpp=[]
plyy=[]
seqq=[]
for i in range(max(df['Clusters'])+1):
    plh=[]
    sbp=[]
    ply=[]
    seq=[]
    for j in range(len(df)):
        if(df['Clusters'][j]==i):
            seq.append(df['Sequence'][j])
            sbp.append(df['Subpopulation'][j])
            plh.append(df['PlantHeight'][j])  
            ply.append(df['Yield'][j])

    a=mean(plh)
    for i in range(len(plh)): plhh.append(a)

    b=mean(ply)
    for i in range(len(ply)): plyy.append(b)

    c=mode(sbp)
    for i in range(len(sbp)): sbpp.append(c)

    d=mode(seq)
    for i in range(len(seq)): seqq.append(d)

            
```

```python
df['Sequence']=seqq
df['Subpopulation']=sbpp
df['PlantHeight']=plhh
df['Yield']=plyy
df

    
```

```python
del df['CultivarID']
del df['Clusters']
```

```python
df.to_csv('clstd.csv')
```
