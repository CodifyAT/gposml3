```python
import pandas as pd
```

```python
df=pd.read_csv('initial files/Genotypicdata.csv')
df.head()
```

```python
prim= df[df['Cultivar ID'] == 'Primary']
prim=prim.values.flatten()
prim
```

```python
secd= df[df['Cultivar ID'] == 'Secondary']
secd=secd.values.flatten()
secd
```

```python
for i in df["Cultivar ID"][6:]:
    a=df[df["Cultivar ID"]==i].values.flatten()
    for j in range(len(a)):
        if(a[j]=='N'):
            a[j]=prim[j]
        elif(a[j]=='DEL'):
            a[j]=secd[j]
    df[df["Cultivar ID"]==i]=a
    
```
