```python
import pandas as pd
```

```python
df=pd.read_csv("gsol2.csv")
df
```

```python
clid=[]
plh=[]
yld=[]
sbp=[]
for i in range(len(df["Cultivar ID"])):
    clid.append(df['Cultivar ID'][i])
    plh.append(df['Plant Height (cm)'][i])
    yld.append(df['Yield'][i])
    sbp.append(df['Subpopulation'][i])
```

```python
del df["Unnamed: 0"]
del df["Plant Height (cm)"]
del df["Yield"]
del df['Subpopulation']
```

```python
seq=[]
for i in df['Cultivar ID']:
    st=""
    a=(df[df['Cultivar ID']==i]).values.flatten()
    for j in a[1:]:
        st+=j
    seq.append(st)
seq
```

```python
data={'Cultivar ID':clid,'Sequences':seq, 'Subpopulation':sbp, 'PlantHeight':plh, 'Yield':yld}
data=pd.DataFrame(data)
data
```

```python
data.to_csv('gsol3.csv')
```
